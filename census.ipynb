# Cuxfilter dashboard for a 310M row Census 2010 dataset

## 1. Import libraries and datasets

```python
import cuxfilter
import cudf
```

```python
#read dataset from disk
df = cudf.read_parquet('/home/ajay/new_dev/plotly/plotly-dash-rapids-census-demo/data/census_data.parquet/*')
#create cuxfilter dataframe
cux_df = cuxfilter.DataFrame.from_dataframe(df)
cux_df.data
```

## 2. Declare charts to be in the dashboard

```python
#declare charts
chart0 = cuxfilter.charts.scatter_geo(x='x', y='y')
chart1 = cuxfilter.charts.bar('age')
chart2 = cuxfilter.charts.multi_select('cow')
chart3 = cuxfilter.charts.bar('income')
chart4 = cuxfilter.charts.bar('education')
```

## 3. Initialize dashboard with various layout and theme options

```python
#intiiate dashboard
d = cux_df.dashboard(
    [chart0, chart1, chart2, chart3, chart4],
    layout=cuxfilter.layouts.feature_and_quad_base,
    theme = cuxfilter.themes.light,
    title='Census Dashboard'
)
```

## 4. Run the dashboard as a web-app

```python
#show the dashboard
notebook_url = 'http://localhost:8888'

d.show()
```
